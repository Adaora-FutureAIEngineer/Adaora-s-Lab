```python
# Hierarchical Clustering Example
# Project: Customer Segmentation using Hierarchical Clustering

# Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.preprocessing import StandardScaler

# Step 2: Create Dataset
data = {
    'Income': [15, 16, 17, 18, 19, 60, 62, 63, 65, 70],
    'Spending_Score': [80, 85, 83, 88, 90, 20, 22, 19, 18, 17]
}
df = pd.DataFrame(data)

# Step 3: Standardize the Data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 4: Perform Hierarchical Clustering
linked = linkage(scaled_data, method='ward')

# Step 5: Create and Visualize Dendrogram
plt.figure(figsize=(8, 5))
dendrogram(linked, labels=df.index.tolist())
plt.title("Customer Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()

# Step 6: Choose number of clusters and assign
num_clusters = 2
df['Cluster'] = fcluster(linked, num_clusters, criterion='maxclust')

# Step 7: Visualize the Clusters
plt.scatter(df['Income'], df['Spending_Score'], c=df['Cluster'], cmap='rainbow')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1–100)")
plt.title("Customer Segmentation using Hierarchical Clustering")
plt.show()

# Step 8: Display Cluster Assignments
print(df)
```
