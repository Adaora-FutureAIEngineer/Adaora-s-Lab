```python
# DBSCAN Example - Customer Location Clustering

# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Step 2: Create example dataset
data = {
    'X': [1, 2, 2, 8, 8, 25, 24, 23, 24, 25],
    'Y': [2, 3, 2, 8, 9, 25, 24, 25, 26, 27]
}
df = pd.DataFrame(data)

# Step 3: Standardize data (DBSCAN works better when features are scaled)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 4: Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=2)
clusters = dbscan.fit_predict(scaled_data)
df['Cluster'] = clusters

# Step 5: Visualize clusters
plt.scatter(df['X'], df['Y'], c=df['Cluster'], cmap='rainbow')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('DBSCAN Clustering Result')
plt.show()

# Step 6: View cluster assignments
print(df)
```
